Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.

Nominal encoding, also known as one-hot encoding, is a technique used in data preprocessing and feature engineering for categorical variables in machine learning. It involves converting categorical data into a binary vector representation where each category is represented by a unique binary column. This technique is particularly useful when dealing with categorical variables that don't have a meaningful ordinal relationship between their values.

Q6. You are working with a dataset containing information about different types of animals, including their
species, habitat, and diet. Which encoding technique would you use to transform the categorical data into
a format suitable for machine learning algorithms? Justify your answer.

We are using One Hot encoding techniques to transform the cateogorical data into  format suitable for machine learning algorithms.

```python
import pandas as pd
```

```python
species = [ 'Loin', 'Elephant', 'Tiger' ]
habitat = [ 'Savanna', 'Jungle', 'Savanna' ]
Diet     = [ 'Carnivore', 'Herbivore', 'Carnivore' ]

df = pd.DataFrame( [ species , habitat , Diet ] , columns= [ 'Species', 'Habitat', 'Diet'] )
```

```python
df.head()
```

```python
## Apply one hot encoding technique 
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform( df[ [ 'Species', 'Habitat', 'Diet' ] ] )
```

```python
pd.DataFrame( encoded.toarray() , columns= encoder.get_feature_names_out() )
```
